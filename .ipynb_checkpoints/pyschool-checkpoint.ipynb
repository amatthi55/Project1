```python
import pandas as pd
import numpy as np 
```

```python
file_one = "Resources/students_complete.csv"
file_two = "Resources/schools_complete.csv"
```

```python
students = pd.read_csv(file_one)
schools =pd.read_csv(file_two)
```

```python
# Sort results in descending order
students = students.sort_values(["reading_score"], ascending=True)
#students.head()
```

```python
"""
Create a high level snapshot (in table form) of the districts key metrics, including:
'Total Schools
'Total Students
'Total Budget
'Average Math Score
'Average Reading Score
'% Passing Math
'% Passing Reading
'Overall Passing Rate (Average of the above two)
"""
```

```python
#Summing Individuals
#Count Number of School
tot_school=schools["name"].count()

#Count Number of Students
Tot_StuNum=students["name"].count()

#Aggregate totals
#Budget
Tot_Bud=schools["budget"].sum()

Avg_mathS=students["math_score"].mean()
Avg_readS=students["reading_score"].mean()


Avg_mathS
#Avg_readS
Tot_Bud
```

```python
#Create a df for those passing
passingR= students.loc[(students["reading_score"] >= 65)]

#count the number of passing
Readp=passingR["name"].count()
percent_passR = ((Readp/Tot_StuNum)*100)

passingM= students.loc[(students["math_score"] >= 65)]
Mathp=passingM["name"].count()
percent_passM = ((Mathp/Tot_StuNum)*100)

#calculates overall percentages
percent_overall = (percent_passR+percent_passM)/2
percent_overall
```

```python
summary_df = pd.DataFrame({"Total Schools":[tot_school],
                          "Total Students":[Tot_StuNum],
                          "Total Budget":[Tot_Bud],
                          "Average Math Score":[Avg_mathS],
                          "Average Reading Score":[Avg_readS],
                          "% Passing Math":[percent_passM],
                          "% Passing Reading":[percent_passR],
                          "Overall Passing Rate":[percent_overall]})
summary_df
```

```python
"""
School Summary


Create an overview table that summarizes key metrics about each school, including:


School Name
School Type
Total Students
Total School Budget
Per Student Budget
Average Math Score
Average Reading Score
% Passing Math
% Passing Reading
Overall Passing Rate (Average of the above two)

"""
```

```python
#changing the name so i can merge the datasets and groupby
schools = schools.rename(columns={"name":"school"})
```

```python
#Merges original data
merge_df = pd.merge(students, schools, how="outer", on="school")
merge_df.head()
merge_df.dtypes
```

```python
school_stats=merge_df.groupby("school")
school_stats.head()
```

```python
#Converts to int
merge_df["reading_score"] = pd.to_numeric(merge_df["reading_score"], errors='coerce').fillna(0).astype(np.int64)
```

```python
# Create DFrame of the different school names
sch_Type = pd.DataFrame(school_stats["type"].value_counts())
sch_Type = sch_Type.rename(columns={"type":"# of Students"})
#sch_Type = sch_Type.reset_index()
sch_Type
```

```python
sch_Type = sch_Type.reset_index()
sch_Type
```

```python

sch_budgets = pd.DataFrame(school_stats["budget"].value_counts())
sch_budgets = sch_budgets.rename(columns={"budget":"delete"})
sch_budgets = sch_budgets.reset_index()
sch_budgets  = sch_budgets.loc[:,["school", "budget"]]
```

```python
#CALCULATES SCHOOL SCORES
sch_scores = school_stats.mean()
sch_scores = sch_scores.rename(columns={"reading_score":"Avg Reading Score","math_score":"Avg Math Score"})
sch_scores = sch_scores.reset_index()
sch_scores["Per Student Budget"] = sch_scores["budget"]/sch_scores["size"]
sch_scores  = sch_scores.iloc[:,[0,2,3,6,7]]
sch_scores
```

```python
#MERGES NEEDED VALUES AND FORMATS
combined_df = pd.merge(sch_Type, sch_scores, how="outer", on="school")
combined_df = combined_df.rename(columns={"school":"School Name", "budget": "School Budget", "type": "Type"})
combined_df
```

```python
#PASSING Reading CALCULATIONS


passing_reading = merge_df.loc[merge_df["reading_score"] >= 65]
#passing_base_df
passing_reading_BySch = passing_reading.groupby("school")
#passing_base_dfBySch.head()
scd_passR = pd.DataFrame(passing_reading_BySch["budget"].value_counts())
#combined_df = combined_df.rename(columns={"school":"School Name", "budget": "School Budget", "type": "Type"})
scd_passR.head()
scd_passR = scd_passR.rename(columns={"budget":"Passing Reading", })
scd_passR = scd_passR.reset_index()
scd_passR["% Passing Reading"] = (scd_passR["Passing Reading"]/combined_df["# of Students"])*100
scd_passR = scd_passR.iloc[:,[0,3]]
scd_passR
```

```python
#PASSING Math CALCULATIONS


passing_math = merge_df.loc[merge_df["math_score"] >= 65]
#passing_base_df
passing_math_BySch = passing_math.groupby("school")
#passing_base_dfBySch.head()
scd_passM = pd.DataFrame(passing_math_BySch["budget"].value_counts())
#combined_df = combined_df.rename(columns={"school":"School Name", "budget": "School Budget", "type": "Type"})
scd_passM.head()
scd_passM = scd_passM.rename(columns={"budget":"Passing Math"})
scd_passM = scd_passM.reset_index()
scd_passM["% Passing Math"] = (scd_passM["Passing Math"]/combined_df["# of Students"])*100
scd_passM = scd_passM.iloc[:,[0,3]]
scd_passM
```

```python
#COMBINES AND REFORMATS
combine_percent = pd.merge(scd_passM, scd_passR, how="outer", on="school")
combine_percent["Overall Pass Rate"]= (combine_percent["% Passing Math"]+combine_percent["% Passing Reading"])/2
combine_percent = combine_percent.rename(columns={"school":"School Name"})
combine_percent
```

```python
#FINAL CHART
combined_df = pd.merge(combined_df, combine_percent, how="right", on="School Name")
combined_df.head()
```
